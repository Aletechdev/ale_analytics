```python
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 200)
```

```python
all_muts_df = pd.read_pickle("../data/ava/4_10_with_uniq_midpts.pkl")
all_muts_df.head()
```

```python
op_df = pd.DataFrame()
for _, m in all_muts_df.iterrows():
    for op in m["operons"]:
        if op["name"] == "pykF":
            op_df = op_df.append(m)
            break
op_df.head()
```

```python
def get_sample_name(exp_name, ale, flask, isolate, tech_rep):
    sample_name = exp_name + " " + str(int(ale)) + " " + str(int(flask)) + " " + str(int(isolate)) + " " + str(int(tech_rep))
    return sample_name


op_df["sample"] = op_df.apply(lambda r: get_sample_name(r.exp, r.ale, r.flask, r.isolate, r.tech_rep), axis=1)
op_df.head()
```

```python
op_df["Mutation Type"].unique()
```

```python
feat_set = set()
for _, m in op_df.iterrows():
    for f in m["genomic features"]:
        feat_set.add(f["name"])
mat = pd.DataFrame(columns=op_df["sample"].unique(), index=feat_set)
mat = mat.fillna('')

for _, m in op_df.iterrows():
    for f in m["genomic features"]:
        curr_mut_type = mat.at[f["name"], m["sample"]]
        if curr_mut_type != m["Mutation Type"]:  # oncoprint doesn't double-count mutation types within the same feature+sample (it might; need to test, but not currently needed), therefore explicitely removing to remember this quirk
            if curr_mut_type != '':
                curr_mut_type += ';'
            new_mut_type = m["Mutation Type"]
            curr_mut_type += new_mut_type
            mat.at[f["name"], m["sample"]] = curr_mut_type

mat.to_csv("./pykF_op_oncoprint_mat_genes.csv")

mat
```

```python
op_df.exp.unique()
```

```python
exp_id_d = {
    "SER": "SER",
    "42C": "42C",
    "GLU": "GLU",
    "SSW_GLU_AC": "SSW_GLU_AC",
    "SSW_GLU_XYL": "SSW_GLU_XYL",
    'TOL_hexanoic_acid': "TOL_hexanoic_acid",
    'TOL_isobutyric_acid': "TOL_isobutyric_acid",
    'TOL_putrescine': "TOL_putrescine",
    "tpi": "tpiAKO",
    "tpiHSA": "tpiHSA",
    "tpiPAE": "tpiPAE",
    "wt": "gentamycin",
    "pyocyanin_supplemented_medium": "pyocyanin",
}


op_df["exp id"] = op_df["exp"].apply(lambda exp_name: exp_id_d[exp_name])
op_df.head()
```

```python
cond_cols = {
    'base-media',
    'calcium-source',
    'carbon-source',
    'nitrogen-source',
    'phosphorous-source',
    'strain-description',
    'sulfur-source',
    'supplement',
    'temperature',
    'exp id'
}
cond_mat = pd.DataFrame(
    columns=op_df["sample"].unique(),
    index=cond_cols
)
cond_mat = cond_mat.fillna('')

for _, m in op_df.iterrows():
    for cond_col in cond_cols:
        cond_mat.at[cond_col, m['sample']] = m[cond_col]
        
# Serine's default supplement annotation is too large.
cond_mat = cond_mat.T
cond_mat["supplement"] = cond_mat["supplement"].apply(lambda x: "glycine(2mM) & L-serine" if x == "glycine(2mM) L-Serine(varying) Wolfe's vitamin solution trace elements(X1)" else x)
cond_mat["supplement"] = cond_mat["supplement"].apply(lambda x: "NaCl(0.5) trace elements" if x == "NaCl(0.5g/L) trace elements" else x)
cond_mat["temperature"] = cond_mat["temperature"].apply(lambda x: x.replace("celsius", "Celsius"))
cond_mat = cond_mat.T
cond_mat.to_csv("./pykF_op_oncoprint_mat_conditions.csv")
cond_mat
```

```python
sel_pres_d = {
    "Herring 2006": "glycerol carbon source",
    "TOL_hexamethylenediamine": "hexamethylenediamine tolerance",
    "GYD": "glycerol carbon source",
    "SSW_GLU_XYL": "glucose/xylose carbon source switching",
    "pts": "ΔptsHIcrr",
    "PGI": "Δpgi",  # Charusanti and Long
    "pgi": "GMOS Δpgi",  # CCK PGI
    "MG1655-M9-NC_000913_3gb-stationary-37-m-tartrate2": "m-tartrate carbon source",
    "42C": "42 Celsius",
    "GLU": "glucose carbon source",
    "SER": "L-serine tolerance",
    "SSW_GLU_AC": "glucose/acetate carbon source switching",
    "TOL_hexanoic_acid": "hexanoic acid tolerance",
    "TOL_isobutyric_acid": "isobutryic acid tolerance",
    "TOL_putrescine": "putrescine tolerance",
    "tpi": "ΔtpiA",
    "pyocyanin_supplemented_medium": "pyocyanin"
}

op_df["selection pressure"] = op_df.apply(lambda m: sel_pres_d[m.exp], axis=1)
```

```python
from collections import Counter
d = Counter(op_df["selection pressure"])
d2 = dict()
for k, v in d.items():
    d2[k] = [v]
cnt_df = pd.DataFrame(d2)
cnt_df = cnt_df.T


import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 300
sns.set_palette("deep")
sns.set_context("paper")
sns.set_style("ticks")
plt.rcParams['font.sans-serif'] = ["Helvetica"]

cnt_df = cnt_df.sort_values(by=0)
ax = cnt_df.plot.barh(
    stacked=True,
#     color=colors,
    figsize=(1, 1.5),
    width=0.90,
)

sns.despine(ax=ax, top=True, right=True, bottom=False, left=True)
ax.tick_params(axis='y', which='major', length=0)
ax.legend(bbox_to_anchor=(1.2, -0.4), prop={"size": 8})

ax.set_title("PykF mutated across\nmultiple experiments", fontname="Helvetica")
# ax.axes.get_xaxis().set_ticks([])
ax.set_xlabel('mutation count', fontsize=8)
ax.set_ylabel('selection pressures', fontsize=8)
# ax.axes.set_title("Title",fontsize=50)
ax.get_legend().remove()
```

```python
targ_exp_l = list(op_df.exp.unique())
targ_exp_l
```

```python
targ_exp_mut_df = all_muts_df[all_muts_df["exp"].isin(targ_exp_l)].copy()
targ_exp_mut_df["exp ale"] = targ_exp_mut_df.apply(lambda r: r["exp"] + ' ' + str(int(r["ale"])), axis=1)
targ_exp_ale_cnt_d = {e:0 for e in targ_exp_mut_df.exp.unique()}
for exp, gdf in targ_exp_mut_df.groupby("exp"):
    targ_exp_ale_cnt_d[exp] = len(gdf["exp ale"].unique())
targ_exp_ale_cnt_d
```

```python
op_df["exp ale"] = op_df.apply(lambda r: r["exp"] + ' ' + str(int(r["ale"])), axis=1)
targ_mut_exp_ale_cnt_d = {e:0 for e in targ_exp_mut_df.exp.unique()}  # want to reuse targ_exp_mut_df from before to ensure full exp coverage between the two set
for exp, gdf in op_df.groupby("exp"):
    targ_mut_exp_ale_cnt_d[exp] = len(gdf["exp ale"].unique())
targ_mut_exp_ale_cnt_d
```

```python
prop_d = dict()
for k in targ_exp_ale_cnt_d.keys():
    prop_d[sel_pres_d[k]] = [targ_mut_exp_ale_cnt_d[k] / targ_exp_ale_cnt_d[k]]  # list for value since easier to make dataframe 
df = pd.DataFrame(prop_d).T
display(df)

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 300
sns.set_palette("deep")
sns.set_context("paper")
sns.set_style("ticks")
plt.rcParams['font.sans-serif'] = ["Helvetica"]

df = df.sort_values(by=0)
ax = df.plot.barh(
    stacked=True,
#     color=colors,
    figsize=(1, 1.8),
    width=0.90,
)

sns.despine(ax=ax, top=True, right=True, bottom=True, left=True)
ax.get_legend().remove()
# plt.xscale('log')
ax.tick_params(axis='y', which='major', length=0)
# ax.get_xaxis().set_visible(False)

df = df.reset_index()  # Resetting index for the below annotations to work
for i, r in df.iterrows():
#     s = "{:.2f}".format(r[0])
    percent = int(r[0] * 100)
    s = str(percent) + '%'
    ax.text(r[0], i-0.35, s, fontname="Helvetica", fontsize=11)

ax.set_title("pykF operon mutated across\nmultiple experiments", fontname="Helvetica")

ax.tick_params(axis='both', which='both', length=0)
ax.set_xticklabels([])
ax.set_xlabel('proportion of ALEs', x=0.7, fontname="Helvetica", fontsize=9)
ax.set_ylabel('', fontname="Helvetica", fontsize="8")
ax.xaxis.labelpad = -2
```
