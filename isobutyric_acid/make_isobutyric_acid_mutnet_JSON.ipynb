```python
import pandas as pd
pd.options.display.max_columns = 100
```

```python
all_mut_df = pd.read_pickle("../data/4_10_with_uniq_midpts.pkl")
display(len(all_mut_df), all_mut_df.head())
```

```python
targ_mut_df = pd.DataFrame()
feat_types = [
    "genomic features",
    "operons",
    "pathways",
#     "regulators",
#     "imodulons",
#     "EC numbers"
]  # ignoring genetic features for now since the genes get double with genomic features

for _, m in all_mut_df.iterrows():
    mut_assoc = False
    for ft in feat_types:
        for f in m[ft]:
            if "isobutyric acid" in f["significantly associated conditions"]:
                mut_assoc = True
                break
        if mut_assoc:
            break

    if mut_assoc:
        targ_mut_df = targ_mut_df.append(m)

display(len(targ_mut_df), targ_mut_df.head())
```

```python
# TODO: don't want to deal with giant deletions since their complicated networks are compounding mutation count issues. Bring them back in later
drop_list_indexes = list(targ_mut_df[targ_mut_df["Sequence Change"]=="Δ18,364 bp"].index)
drop_list_indexes += list(targ_mut_df[targ_mut_df["Sequence Change"]=="Δ40,006 bp"].index)
targ_mut_df = targ_mut_df.drop(drop_list_indexes)
```

```python
geno_feat_tot_mut_cnts = dict()
gen_feat_tot_mut_cnts = dict()
op_tot_mut_cnts = dict()
pwy_tot_mut_cnts = dict()
react_tot_mut_cnts = dict()
reg_tot_mut_cnts = dict()
imod_tot_mut_cnts = dict()
# Using the individual feature columns rather than link columns because this guarantees considering all linked and non-linked features
# for each of these levels, except genomic features.
for _, m in targ_mut_df.iterrows():
    
    for geno_feat in m["genomic features"]:
        geno_feat_annot = geno_feat["RegulonDB ID"] #+ ' ' + r["Mutation Type"]
        if geno_feat_annot not in geno_feat_tot_mut_cnts.keys():
            geno_feat_tot_mut_cnts[geno_feat_annot] = 0
        geno_feat_tot_mut_cnts[geno_feat_annot] += 1

    for f, links in m["genetic feature links"].items():
        if f not in gen_feat_tot_mut_cnts.keys():
            gen_feat_tot_mut_cnts[f] = 0
        gen_feat_tot_mut_cnts[f] += len(links)

    for f, links in m["operon links"].items():
        if f not in op_tot_mut_cnts.keys():
            op_tot_mut_cnts[f] = 0
        op_tot_mut_cnts[f] += len(links)
        
    for f, links in m["pathway links"].items():        
        if f not in pwy_tot_mut_cnts.keys():
            pwy_tot_mut_cnts[f] = 0
        pwy_tot_mut_cnts[f] += len(links)
        
#     for f, links in m["regulator links"].items():     
#         if f not in reg_tot_mut_cnts.keys():
#             reg_tot_mut_cnts[f] = 0
#         reg_tot_mut_cnts[f] += len(links)
        
#     for f, links in m["imodulon links"].items():     
#         if f not in imod_tot_mut_cnts.keys():
#             imod_tot_mut_cnts[f] = 0
#         imod_tot_mut_cnts[f] += len(links)
        
#     for f, links in m["EC number links"].items():     
#         if f not in react_tot_mut_cnts.keys():
#             react_tot_mut_cnts[f] = 0
#         react_tot_mut_cnts[f] += len(links)
```

```python
FEAT_COLS = [
    "genomic features",
#     "genetic features",
    "operons",
    "pathways",
#     "regulators",
#     "imodulons"
]

node_d = dict()  # don't want duplicate node entries, therefore going to use dictionary test enforce uniqueness
for _, m in targ_mut_df.iterrows():
    for c in FEAT_COLS:
        for f in m[c]:
            
            mut_cnt = 0
            name = f["name"]
            if c == "genomic features":
                mut_cnt = geno_feat_tot_mut_cnts[f["RegulonDB ID"]]
#                 name += " genomic feature"
#             if c == "genetic features":
#                 mut_cnt = gen_feat_tot_mut_cnts[f["RegulonDB ID"]]
#                 name += " genetic feature"
            if c == "operons":
                mut_cnt = op_tot_mut_cnts[f["RegulonDB ID"]]
                name += " operon"
            if c == "pathways":
                mut_cnt = pwy_tot_mut_cnts[f["name"]]
                name += " pathway"
#             if c == "regulators":
#                 mut_cnt = reg_tot_mut_cnts[f["name"]]
#                 name += " regulon"
#             if c == "imodulons":
#                 mut_cnt = imod_tot_mut_cnts[f["name"]]
#                 name += " imodulon"
                
            # The id need to match what's in the link dict
            id_s = name  # Used for high-level functions (regs, pwys, etc.)
            if "RegulonDB ID" in f.keys():
                
                id_s = f["RegulonDB ID"]
#                 if c == "genomic features": id_s += " genomic feature"
#                 if c == "genetic features": id_s += " genetic feature"
                if c == "operons": id_s += " operon"  # though don't need to differentiate, like with geno and gen, easier to debug
#                 if c == "regulators": id_s += " regulon"  # though don't need to differentiate, like with geno and gen, easier to debug
#                 if c == "imodulons": id_s += " imodulon"
                
            if id_s not in node_d.keys():
                node_d[id_s] = {"id": id_s, "name": str(mut_cnt) + " " + name, "count":mut_cnt}

nodes = []
for k, v in node_d.items():
    nodes.append({"data": v})
```

```python
# The below code links genomic features directly to operons, skipping genetic features
FEAT_LINK_COLS = [
    'operon links',
    'pathway links',
]


edge_d = dict()  # The issue here is I don't want duplicate edges within the dataset; therefore using a dict check for duplicates

for _, m in targ_mut_df.iterrows():            
    for c in FEAT_LINK_COLS:
        
        link_d = m[c]
        
        if c == "operon links":
            for k, l in link_d.items():
                
                for v in set(l):  # Must only look at unique entries in the case of overlap features
                    target_id = k + " operon"
                    geno_sources = m["genetic feature links"][v]
                    
                    for source_id in geno_sources:
                        id_s = target_id + '-' + source_id
                        if id_s not in edge_d.keys():
                            edge_d[id_s] = {"id": id_s, 'target': target_id, "source": source_id, "count": 1}
                        else:
                            edge_d[id_s]["count"] += 1  # Figure out where to put this count line to fix bug
        
        if c == "pathway links":
            for k, l in link_d.items():
                for v in l:

                    target_id = k + " pathway"
                    source_id = v + " operon"
                    
                    id_s = target_id + '-' + source_id
                    if id_s not in edge_d.keys():
                        edge_d[id_s] = {"id": id_s, 'target': target_id, "source": source_id, "count": 1}
                    else:
                        edge_d[id_s]["count"] += 1
                    
edges = []
for k, v in edge_d.items():
    edges.append({"data":v})
```

```python
mutnet_data = dict()
mutnet_data["nodes"] = nodes
mutnet_data["edges"] = edges
```

```python
import json
with open('isobutyric_acid_mutnet.json', 'w') as outfile:
    json.dump(mutnet_data, outfile)
```
